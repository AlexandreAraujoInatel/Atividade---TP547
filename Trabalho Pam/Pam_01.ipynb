```python
import numpy as np  # Importa o NumPy para operações numéricas
import matplotlib.pyplot as plt  # Importa o Matplotlib para visualização gráfica

# Número total de símbolos que queremos gerar
num_symbols = 100  # Isso será o comprimento do sinal 4-PAM

# Criar um vetor de símbolos aleatórios (cada símbolo é representado por 2 bits, ou seja, um valor entre 0 e 3)
# np.random.choice() seleciona aleatoriamente de um conjunto de valores
symbols = np.random.choice([0, 1, 2, 3], size=num_symbols)  # Cria uma sequência de 100 símbolos aleatórios

# Um dicionário para mapear cada valor de símbolo para sua respectiva amplitude no esquema 4-PAM
# Neste esquema, associar cada valor de 0 a 3 a uma amplitude distinta
pam_map = {
    0: -3,  # 0 é mapeado para amplitude -3
    1: -1,  # 1 é mapeado para amplitude -1
    2: 1,   # 2 é mapeado para amplitude 1
    3: 3,   # 3 é mapeado para amplitude 3
}

# Usar uma compreensão de lista para converter os símbolos em amplitudes 4-PAM
# Cada símbolo é substituído por sua amplitude correspondente usando o dicionário pam_map
pam_signal = np.array([pam_map[s] for s in symbols])

# Visualização do sinal 4-PAM gerado
plt.figure(figsize=(12, 6))  # Configura o tamanho da figura para a visualização
plt.plot(pam_signal, 'bo-', label='Sinal 4-PAM')  # Plota o sinal como uma linha com pontos azuis
plt.title("Sinal 4-PAM")  # Adiciona um título ao gráfico
plt.xlabel("Índice")  # Nomeia o eixo x (representa a posição do símbolo)
plt.ylabel("Amplitude")  # Nomeia o eixo y (representa a amplitude do sinal)
plt.grid(True)  # Adiciona uma grade para melhor visualização
plt.legend()  # Adiciona uma legenda para identificar o sinal
plt.show()  # Exibe o gráfico

# Visualizar a distribuição de amplitudes para entender como estão distribuídas
plt.figure(figsize=(12, 6))  # Configura o tamanho da figura para o histograma
# Plota um histograma para mostrar a frequência de cada amplitude
plt.hist(pam_signal, bins=np.arange(-4, 5, 2), rwidth=0.8, align='mid', color='skyblue', edgecolor='black')
plt.title("Distribuição de Amplitudes do Sinal 4-PAM")  # Adiciona um título ao histograma
plt.xlabel("Amplitude")  # Nomeia o eixo x do histograma (representa as amplitudes do 4-PAM)
plt.ylabel("Frequência")  # Nomeia o eixo y do histograma (representa a frequência de cada amplitude)
plt.grid(True)  # Adiciona uma grade ao histograma para melhor visualização
plt.show()  # Exibe o histograma
```
