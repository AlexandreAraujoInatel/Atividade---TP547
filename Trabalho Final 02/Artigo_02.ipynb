```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
fs = 1000  # Frequência de amostragem
f_carrier = 100  # Frequência da portadora
t = np.arange(0, 1, 1/fs)  # Vetor de tempo para 1 segundo de duração com passos de 1/fs
data = np.random.randint(0, 2, int(fs/10))  # Gera dados binários aleatórios (0 ou 1)

# Repetir cada bit para cobrir a janela de tempo da portadora
data_repeated = np.repeat(data, fs//len(data))

# Gera o sinal da portadora (sinal senoidal)
carrier = np.sin(2 * np.pi * f_carrier * t)

# Modulação ASK
ask_signal = carrier * data_repeated  # Multiplica a portadora pelos dados binários

# Modulação FSK
f_carrier1 = 80  # Frequência para representar o bit 0
f_carrier2 = 120  # Frequência para representar o bit 1
# Gera o sinal FSK alternando entre duas frequências de portadora
fsk_signal = (np.sin(2 * np.pi * f_carrier1 * t) * (data_repeated == 0) +
              np.sin(2 * np.pi * f_carrier2 * t) * (data_repeated == 1))

# Modulação PSK
# Altera a fase da portadora de acordo com os dados binários
psk_signal = np.sin(2 * np.pi * f_carrier * t + np.pi * data_repeated)

# Plotagem dos sinais
plt.figure(figsize=(12, 8))

# Sinal de Dados Original
plt.subplot(4, 1, 1)
plt.title('Data Signal')
plt.plot(t, data_repeated, 'k')  # Plota os dados binários repetidos
plt.ylim(-1.5, 1.5)  # Define os limites do eixo y

# Sinal ASK
plt.subplot(4, 1, 2)
plt.title('ASK Signal')
plt.plot(t, ask_signal)  # Plota o sinal modulado em ASK
plt.ylim(-1.5, 1.5)  # Define os limites do eixo y

# Sinal FSK
plt.subplot(4, 1, 3)
plt.title('FSK Signal')
plt.plot(t, fsk_signal)  # Plota o sinal modulado em FSK
plt.ylim(-1.5, 1.5)  # Define os limites do eixo y

# Sinal PSK
plt.subplot(4, 1, 4)
plt.title('PSK Signal')
plt.plot(t, psk_signal)  # Plota o sinal modulado em PSK
plt.ylim(-1.5, 1.5)  # Define os limites do eixo y

plt.tight_layout()  # Ajusta o layout para não sobrepor os gráficos
plt.show()  # Exibe os gráficos

```
